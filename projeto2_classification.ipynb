**Utilities**

Introduziremos uma função para reduzir a memória utilizada pelo nosso dataset ao escolher o menor tipo em tamanho de bytes capaz de armazenar todos os valores de uma determinada coluna.

```
# Reduce memory usage
def get_optimized_column(column):
    if column.dtype.name == 'category':
      return column
    if not np.issubdtype(column.dtypes, np.number):
        return column
    integers = [np.int8, np.int16, np.int32, np.int64]
    floats = [np.float16, np.float32, np.float64]
    max = column.max()
    relevant_types = integers if np.issubdtype(column.dtypes, np.integer) else floats
    for dtype in relevant_types:
        if dtype(max) == max:
            return column.astype(dtype)
    return column


def reduce_mem_usage(df, verbose=True):
    start_mem = df.memory_usage().sum() / 1024**2    
    for column in df.columns:
        df[column] = get_optimized_column(df[column]) 
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

**Vamos substituir os valors nulos das colunas 'AvgH', 'AvgD', 'AvgA'**

1X2 (match) betting odds data:
(Colunas dos dados das casas de aposta)

Home Win odds columns: B365H, BSH, BWH, GBH, IWH, LBH, PSH, PH, SOH, SBH, SJH, SYH, VCH, WHH

Draw odds columns: 'B365D', 'BSD', 'BWD', 'GBD', 'IWD', 'LBD', 'PSD', 'PD', 'SOD', 'SBD', 'SJD', 'SYD', 'VCD', 'WHD'

Away Win odds columns: B365H, BSH, BWH, GBH, IWH, LBH, PSH, PH, SOH, SBH, SJH, SYH, VCH, WHH

```
#Obtendo as médias das linhas que nós queremos
cols_home_win_odds_mean = np.nanmean(full_soccer_data[['B365H', 'BSH', 'BWH', 'GBH', 'IWH', 'LBH', 'PSH', 'PH', 'SOH', 'SBH', 'SJH', 'SYH', 'VCH', 'WHH']], axis=1) #verificar isso: quero pegar a média de cada  LINHA
cols_draw_odds_mean = np.nanmean(full_soccer_data[['B365D', 'BSD', 'BWD', 'GBD', 'IWD', 'LBD', 'PSD', 'PD', 'SOD', 'SBD', 'SJD', 'SYD', 'VCD', 'WHD']], axis=1)
cols_away_win_odds_mean = np.nanmean(full_soccer_data[['B365A', 'BSA', 'BWA', 'GBA', 'IWA', 'LBA', 'PSA', 'PA', 'SOA', 'SBA', 'SJA', 'SYA', 'VCA', 'WHA']], axis=1)

#Checando quantos valores nulos em cada numpy array com os dados que usaremos para substituir
#print(np.count_nonzero(np.isnan(cols_home_win_odds_mean)))
#print(np.count_nonzero(np.isnan(cols_draw_odds_mean)))
#print(np.count_nonzero(np.isnan(cols_away_win_odds_mean)))

#Checando quais indices precisamos substituir
inds_H = np.where(np.isnan(full_soccer_data[['AvgH']]))
inds_D = np.where(np.isnan(full_soccer_data[['AvgD']]))
inds_A = np.where(np.isnan(full_soccer_data[['AvgA']]))

#Colcando as médias das colunas nos índices (do dataframe soccer_data). Alinhando os arrays usando take
full_AvgH_to_replace = full_soccer_data[['AvgH']].to_numpy()
full_AvgH_to_replace[inds_H] = np.take(cols_home_win_odds_mean, inds_H[0])
soccer_data[['AvgH']] = full_AvgH_to_replace
#----#
full_AvgD_to_replace = full_soccer_data[['AvgD']].to_numpy()
full_AvgD_to_replace[inds_D] = np.take(cols_draw_odds_mean, inds_D[0])
soccer_data[['AvgD']] = full_AvgD_to_replace
#----#
full_AvgA_to_replace = full_soccer_data[['AvgA']].to_numpy()
full_AvgA_to_replace[inds_A] = np.take(cols_away_win_odds_mean, inds_A[0])
soccer_data[['AvgA']] = full_AvgA_to_replace

print(soccer_data[['AvgH']].isnull().sum())
print(soccer_data[['AvgA']].isnull().sum())
print(soccer_data[['AvgD']].isnull().sum())
```

# Hyperparameter optimization

Antes de tudo, vamos instalar e configurar o MLFlow para poder rastrear os melhores modelos utilizados

```
import mlflow
from urllib.parse import urlparse
```

---
Vamos utilizar o [Auto-sklearn](https://automl.github.io/auto-sklearn/master/) para selecionar o melhor modelo e realizar tuning dos hyperparametros para os modelos sklearn.



Para o **XGBoost**, que não pertence a biblioteca sklearn, e portanto, não pode ser utilizado com o autosklearn, nós iremos utilizar a biblioteca [HyperOpt](http://hyperopt.github.io/hyperopt/). HyperOpt precisa de uma função objetivo e um espaço de busca, iremos utilizar parâmetros próximos do que temos por padrão e verificar quais seriam os melhores valores para esse vetor de parâmetros utilizando a função fmin para minimizar nossa função objetivo.


 - Primeiro, vamos fazer para o **XGBoost**


```
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials

#Criando o conjunto de validação
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, random_state=seed, test_size=0.2, stratify=y_train)

def objective(space):
    
    with mlflow.start_run():
      clf = xgb.XGBClassifier(n_estimators = 4000,
                              max_depth = int(space['max_depth']),
                              min_child_weight = space['min_child_weight'],
                              subsample = space['subsample'],
                              gamma=space['gamma'],
                              colsample_bytree=space['colsample_bytree'],
                              objective= 'binary:logistic',
                              n_jobs=-1,
                              random_state =seed)

      eval_set  = [( X_train, y_train), ( X_valid, y_valid)]

      clf.fit(X_train, y_train,
              eval_set=eval_set, eval_metric="auc",
              early_stopping_rounds=100, verbose=0)

      pred = clf.predict_proba(X_valid)[:,1]
      auc = roc_auc_score(y_valid, pred)
      print("SCORE:", auc)

      mlflow.log_param("max_depth", space['max_depth'])
      mlflow.log_param("min_child_weight", space['min_child_weight'])
      mlflow.log_param("subsample", space['subsample'])
      mlflow.log_param("gamma", space['gamma'])
      mlflow.log_param("colsample_bytree", space['colsample_bytree'])
      mlflow.log_metric("auc", auc)
      tracking_url_type_store = urlparse(mlflow.get_tracking_uri()).scheme
      if tracking_url_type_store != "file":
        mlflow.sklearn.log_model(clf, "model", registered_model_name="XGBoostModel")
      else:
        mlflow.sklearn.log_model(clf, "model")
    return{'loss':1-auc, 'status': STATUS_OK }

space = {
    'eta': hp.uniform('eta', 0.025, 0.5),
    'max_depth': hp.choice('max_depth', range(3, 12)),
    'min_child_weight': hp.quniform ('min_child_weight', 1, 10, 1),
    'subsample': hp.uniform ('subsample', 0.5, 1),
    'gamma': hp.uniform('gamma', 0.1, 1),
    'colsample_bytree': hp.quniform('colsample_bytree', 0.5, 1, 0.05)
}


trials = Trials()
best = fmin(fn=objective, space=space, algo=tpe.suggest,
            max_evals=2,
            trials=trials,
            verbose = 0)

print(best)
```

Modelo XGBoost selecionado com os seguintes parâmetros:


'colsample_bytree': 0.5,

'eta': 0.1684059515028448, 

'gamma': 0.3135452645070162, 

'max_depth': 1, 

'min_child_weight': 9.0, 

'subsample': 0.6805158891485817

---

Vamos treinar o modelo com os parâmetros selecionados e checar o desempenho.

```
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, random_state=seed, test_size=0.2, stratify=y_train)

predictions_list = np.zeros([len(y_train), 2])
predictions_test = np.zeros([len(X_test)])

folds = StratifiedKFold(n_splits=7, shuffle=False, random_state=seed)

for train_idx, valid_idx in folds.split(X_train, y_train):
  X_train_splt, X_valid_splt = X_train.iloc[train_idx,:], X_train.iloc[valid_idx,:]
  y_train_splt, y_valid_splt = y_train.iloc[train_idx], y_train.iloc[valid_idx]

  clf = xgb.XGBClassifier(n_estimators = 4000,
                          objective= 'binary:logistic',
                          n_jobs=-1,
                          eta = best['eta'],
                          max_depth=int(best['max_depth']),
                          min_child_weight=best['min_child_weight'],
                          subsample=best['subsample'],
                          gamma=best['gamma'],
                          colsample_bytree=best['colsample_bytree'],
                          random_state =seed)

  clf.fit(X_train_splt, y_train_splt, eval_set=[(X_train_splt, y_train_splt), (X_valid_splt, y_valid_splt)], 
                early_stopping_rounds=100, eval_metric='auc', verbose=0)
  
  clf.predict(X_valid)
  predictions_list = clf.predict_proba(X_valid) / 2
  predictions_test += clf.predict_proba(X_test[X_valid.columns])[:,1]/ 2

predictions = np.argmax(predictions_list, axis=1)

print(classification_report(y_valid, predictions))
```

Obtivemos uma acurácia não tão boa de 0.67; Vamos calcular o ROC_AUC, nossa métrica principal, e realizar alguns plots.

```
print("CV roc_auc_score", roc_auc_score(y_valid, predictions_list[:,1], average = "macro"))

sns.set(rc={'figure.figsize':(8,8)})
subplot = skplt.metrics.plot_confusion_matrix(y_valid, predictions, normalize=True)
subplot.set_ylim(-0.5, 1.5)

sns.set(rc={'figure.figsize':(8,8)})
skplt.metrics.plot_roc(y_valid, predictions_list)

sns.set(rc={'figure.figsize':(8,8)})
skplt.metrics.plot_lift_curve(y_valid, predictions_list)

sns.set(rc={'figure.figsize':(12, 38)})
xgb.plot_importance(clf, title='Feature importance', xlabel='F score', ylabel='Features')
```
